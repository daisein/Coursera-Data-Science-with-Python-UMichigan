```python
import pandas as pd

doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

df = pd.Series(doc)
df.head(20)
```

```python
all_indexes = []
all_dates = []
```

```python
import re

typeOneDates = []
typeOneIdx = -1

for string in df:
    date = re.findall(r'\d{1,2}[/-]\d{1,2}[/-]\d{2,4}', string)
    typeOneIdx += 1
    if len(date) > 0:
        typeOneDates.append((typeOneIdx, date))
```

```python
typeOneDates
```

```python
#type two: July 11, 1997

typeTwoDates = []
typeTwoIdx = -1
```

```python
for string in df:
    date = re.findall(r'(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (?:\d{2}, )?\d{4}', string)
    typeTwoIdx += 1
    if len(date) > 0:
        typeTwoDates.append((typeTwoIdx, date))
```

```python
typeTwoDates
```

```python
#type three: 6/1998
typeThreeDates = []
typeThreeIdx = -1

for string in df:
    date = re.findall(r'\d{1,2}[/-]\d{4}', string)
    typeThreeIdx += 1
    if len(date) > 0 and typeThreeIdx>342:
        typeThreeDates.append((typeThreeIdx, date))
```

```python
typeThreeDates
```

```python
#type four: 1998
typeFourDates = []
typeFourIdx = -1

for string in df:
    date = re.findall(r'\d{4}', string)
    typeFourIdx += 1
    if len(date) > 0 and typeFourIdx>454:
        typeFourDates.append((typeFourIdx, date))
```

```python
typeFourDates
```

```python
for dateOne in typeOneDates:
    if '/' in dateOne[1][0]:
        month_day_year = dateOne[1][0].split('/')
    elif '-' in dateOne[1][0]:
        month_day_year = dateOne[1][0].split('-')
    #print(month_day_year)
    
    if len(month_day_year[0]) == 1:
        monthOne = '0' + month_day_year[0]
    else:
        monthOne = month_day_year[0]
    #print(monthOne)
    
    if len(month_day_year[1]) == 1:
        dayOne = '0' + month_day_year[1]
    else:
        dayOne = month_day_year[1]
    #print(dayOne)

    if len(month_day_year[2]) == 2:
        if int(month_day_year[2]) > 30:
            yearOne = '19' + month_day_year[2]
        elif int(month_day_year[2]) <30:
            yearOne = '20' + month_day_year[2]
    else:
        yearOne = month_day_year[2]
    #print(yearOne)
    print(dateOne[0], yearOne+monthOne+dayOne)
    all_indexes.append(dateOne[0])
    all_dates.append(yearOne+monthOne+dayOne)
```

```python
monstrToMonint= {'Jan': '01', 'Janaury': '01','January':'01', 'Feb':'02', 'February':'02', 'Mar':'03', 'March':'03', 
                 'April':'04', 'Apr':'04', 'May':'05', 
               'June':'06', 'Jun':'06', 'July':'07', 'Jul':'07', 'August':'08', 'Aug':'08', 'September':'09', 'Sep':'09', 
                 'October':'10', 'Oct':'10', 'November':'11', 'Nov':'11', 'December':'12', 'Dec':'12', 'Decemeber':'12'}
```

```python
for dateTwo in typeTwoDates:
    if dateTwo[0] < 194:
        #print(dateTwo[1][0].split(' '))
        day_month_year = dateTwo[1][0].split(' ')
        
        dayTwo = day_month_year[0]
        #print(dayTwo)

        monthTwo = monstrToMonint[day_month_year[1]]
        #print(monthTwo)
        
        yearTwo = day_month_year[2]
        #print(yearTwo)
        
        print(dateTwo[0], yearTwo+monthTwo+dayTwo)
        all_indexes.append(dateTwo[0])
        all_dates.append(yearTwo+monthTwo+dayTwo)
```

```python
for dateTwo in typeTwoDates:
    if 193 < dateTwo[0] < 228:
        month_day_year = dateTwo[1][0].split(' ')
        #print(month_day_year)
        
        dayTwo = month_day_year[1][:-1]
        #print(dayTwo)

        monthTwo = monstrToMonint[month_day_year[0]]
        #print(monthTwo)
        
        yearTwo = month_day_year[2]
        #print(yearTwo)
        
        print(dateTwo[0], yearTwo+monthTwo+dayTwo)
        all_indexes.append(dateTwo[0])
        all_dates.append(yearTwo+monthTwo+dayTwo)
```

```python
for dateTwo in typeTwoDates:
    if dateTwo[0] > 226:
        month_day_year = dateTwo[1][0].split(' ')
        #print(month_day_year)
        
        dayTwo = '01'
        #print(dayTwo)

        monthTwo = monstrToMonint[month_day_year[0]]
        #print(monthTwo)
        
        yearTwo = month_day_year[1]
        #print(yearTwo)
        
        print(dateTwo[0], yearTwo+monthTwo+dayTwo)
        all_indexes.append(dateTwo[0])
        all_dates.append(yearTwo+monthTwo+dayTwo)
```

```python
for dateThree in typeThreeDates:
    month_day_year = dateThree[1][0].split('/')
    #print(month_day_year)

    dayThree = '01'
    #print(dayTwo)
    
    if len(month_day_year[0]) < 2:
        monthThree = '0' + month_day_year[0]
    else:
        monthThree = month_day_year[0]
    #print(monthTwo)

    yearThree = month_day_year[1]
    #print(yearTwo)

    print(dateThree[0], yearThree+monthThree+dayThree)
    all_indexes.append(dateThree[0])
    all_dates.append(yearThree+monthThree+dayThree)
```

```python
for dateFour in typeFourDates:   
    dayFour = '01'
    monthFour = '01'
    yearFour = dateFour[1][0]
    print(dateFour[0], yearFour+monthFour+dayFour)
    all_indexes.append(dateFour[0])
    all_dates.append(yearFour+monthFour+dayFour)
```

```python
comparison_ls = []
for num in range(0, 500):
    comparison_ls.append(num)
comparison_ls = list(set(comparison_ls) - set(all_indexes))
comparison_ls.sort()
comparison_ls
```

```python
comparison_dates= ['20131011', '19860124', '19760310','19901023', '20040812', 
 '19740910','19751214','19741014','20110915','19900701','20110501','20130101','20060501',
 '20100201','20040501','19950201','19750301','19990602','20010501','20000301','19980401','20050301']
```

```python
for idx in comparison_ls:
    all_indexes.insert(idx, idx)
```

```python
for idx, elem in zip(comparison_ls,comparison_dates):
    all_dates.insert(idx, elem)
```

```python
dates = pd.Series(all_dates)
dates.sort_values(inplace=True)
ans = pd.Series(dates.index)
```
