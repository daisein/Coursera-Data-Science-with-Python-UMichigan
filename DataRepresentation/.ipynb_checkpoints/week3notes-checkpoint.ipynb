# Histograms

```python
fig, ((axi, ax2), (ax3, ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1, ax2, ax3, ax4]
for n in range(0,len(axs)):
    sample_size = 10 ** (n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs=[ax1, ax2, ax3, ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample,bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

# Box and Whisker Plots

```python
# build a df
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample,
                   'random': random_sample,
                   'gamma': gamma_sample})
df
```

```python
#introducing describe()
df.describe()
```

```python
#create a boxplot
plt.figure()
_ = plt.boxplot(df['normal'], whis='range')
```

```python
#boxplots for all 3 columns in df
plt.clf()
_ = plt.boxplot([ df['normal'], df['random'], df['gamma'] ], whis='range')
```

```python
#histogram figure for df[gamma] with 100bins
plt.figure()
_ = plt.hist(df['gamma'], bins=100)
```

```python
#overlay axis, plot the histogram figure in the boxplot
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([ df['normal'], df['random'], df['gamma']], whis='range')
ax2 = mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.5)
```

```python
#switch tick position
ax2.yaxis.tick_right()
```

```python
#bad result happens when you dont pass 'whis'
plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']])
```

# Animations

```python
# prep work for animation
import matplotlib.animation as animation
n = 100
x = np.random.randn(n)
```

```python
#function that animation calls, stops when curr equals the last frame
def update(curr):
    if curr == n: 
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [3,27])
```

```python
fig=plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
```

# Interactivity

```python
#plot a chart and add a clickevent
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixels {}, {} \nand data {}, {}'.format(event.x, event.y, event.xdata, event.ydata))

plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
# make a df
from random import shuffle
origins = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']
shuffle(origins)
df = pd.DataFrame({'height': np.random.rand(10),
                   'weight': np.random.rand(10),
                   'origin': origins
                  })
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))
    
plt.gcf().canvas.mpl_connect('pick event', onpick)
```
