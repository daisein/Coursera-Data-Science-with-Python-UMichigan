```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import GradientBoostingClassifier

train_data = pd.read_csv('train.csv',encoding = 'ISO-8859-1', low_memory=False)
train_data = train_data[pd.notnull(train_data.compliance)]
train_data
```

```python
features = ['fine_amount',
       'admin_fee', 'state_fee', 'late_fee', 'discount_amount',
       'clean_up_cost', 'judgment_amount']
target = ['compliance']

X_train, X_test, y_train, y_test = train_test_split(train_data[features], train_data[target], random_state=0)

nnclf = GradientBoostingClassifier().fit(X_train, y_train)
nnclf.score(X_train, y_train)
```

```python
test_data = pd.read_csv('test.csv')
test_data.set_index('ticket_id')
test_data
```

```python
y_pred = nnclf.predict_proba(test_data[features])
y_pred
```

```python
test_data['compliance'] = y_pred[:,1]
test_data = test_data.round(4)
```

```python
roc_auc_score(y_test,nnclf.predict_proba(X_test)[:,1],)
```
